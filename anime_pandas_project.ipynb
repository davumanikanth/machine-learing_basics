```python
import re

def extract_time(txt):
    # Regex to find patterns like "Month Year - Month Year" or "Month Year"
    # located after an episode count (e.g., "(XX eps)")
    match = re.search(r'\(\d+\s*eps\)([A-Za-z]{3}\s\d{4}(?:\s-\s[A-Za-z]{3}\s\d{4})?)', txt)
    if match:
        return match.group(1).strip()

    # For movie entries, sometimes it's "Movie (1 eps)Sep 2016 - Sep 2016"
    # Also look for date patterns after "Movie (X eps)" if the first regex doesn't catch it.
    match_movie = re.search(r'Movie\s\(\d+\s*eps\)([A-Za-z]{3}\s\d{4}(?:\s-\s[A-Za-z]{3}\s\d{4})?)', txt)
    if match_movie:
        return match_movie.group(1).strip()

    return None # Return None if no timestamp is found
```

```python
df['Timestamp']=df['Title'].apply(extract_time)
```

```python
def calculate_duration(ts):
    ts = ts.replace('.', '')
    start, end = ts.split(' - ')

    start_date = pd.to_datetime(start, format="%b %Y")
    end_date = pd.to_datetime(end, format="%b %Y")

    return (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
```

```python
df["Duration_Months"] = df["Timestamp"].apply(calculate_duration)


print(df)
```
